```python
import pandas as pd
import numpy as np
df = pd.read_csv(r"D:\Internship Program\Multi-Domain Data Analysis Portfolio\Supermarket Sales Analysis\data\supermarket_sales.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.dtypes
```

```python
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Month_Name'] = df['Date'].dt.month_name()
df['Day_Name'] = df['Date'].dt.day_name()
df.dtypes
```

```python
# Time column ko sirf time format me convert karo
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M').dt.time
df.dtypes
```

```python
df.groupby('Product_Line')['Total'].sum()
```

```python
df.groupby('Branch')['Total'].mean()
```

```python
df['Payment'].value_counts()
```

```python
import matplotlib.pyplot as plt

df.groupby('Product_Line')['Total'].sum().plot(kind='bar')
plt.title('Total Sales by Product_Line')
plt.xlabel('Product_Line')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
df.groupby('Branch')['Total'].sum().plot(kind='bar')
plt.title('Branch-wise Total Sales')
plt.xlabel('Branch')
plt.ylabel('Total Sales')
plt.show()
```

```python
df['Payment'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Payment Method Distribution')
plt.ylabel('')
plt.show()
```

```python
df.groupby('Month_Name')['Total'].sum().plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

num_cols = ['Unit_Price', 'Quantity', 'Tax', 'Total', 'Rating']
corr = df[num_cols].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
